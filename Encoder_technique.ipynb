##Encoding Techniques

1)Label Encoding
2)One-Hot Encoding
3)Ordinal Encoding
4)Target Encoding

```python
#Label Encoding-->It is used to encode categarical col to num because algorthim understand only numbers.
'''
#In  this we encode each  value to number from 0 to n.

                 syntax:from sklearn.preprocessing import LabelEncoder

we  use for target column because it has a less unique values for classification algorithm like yes\no,pass/fail

Example : we have target col like PASS/FAIL

                  1)Take unique character
                              PASS
                              FAIL
                  2)Arange in Alphabet order

                            FAIL
                            PASS
                  3)Assign values

                         FAIL----0
                         PASS----1
                         
'''

```

```python
#import lib

import pandas as pd
import numpy as np
n_rows=10
data={
    "Result":np.random.choice(["PASS","FAIL"],n_rows)
}

df=pd.DataFrame(data)
```

```python
df
```

```python
#import LabelEncoder
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["Result"]=le.fit_transform(df["Result"])
df
```

```python
#2 -One hot encoder

'''   
--->It is encoding technique which create separate columns to unique col
'''
```

```python
#import  OneHotEncoder
n_rows=10
data={
    "Result":np.random.choice(["GOOD","BAD","AVERAGE"],n_rows)
}

d=pd.DataFrame(data)

from sklearn.preprocessing import OneHotEncoder

ohe=OneHotEncoder(sparse_output=False)

values=ohe.fit_transform(d)
col=ohe.get_feature_names_out()

d[col]=values
```

```python
d
```

```python
#ordinalEncoder
#It is used when output is in meaning full arder like poor,avg,good   and  small/large/extra large
n_row=10
data={
    "Size":np.random.choice(["SMALL","MEDIAM","LARGE","EXTRA LARGE"],n_row),
    "Result":np.random.choice(["BAD","AVG","GOOD"],n_row)
}
f=pd.DataFrame(data)

from sklearn.preprocessing import OrdinalEncoder

oe=OrdinalEncoder(categories=[
        ["SMALL", "MEDIAM", "LARGE", "EXTRA LARGE"],
        ["BAD", "AVG", "GOOD"]])
df[["Size","Result"]]=oe.fit_transform(f[["Size","Result"]])
```
