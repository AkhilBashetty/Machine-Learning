Advantages:

1)Easy to understand
2)Easy to implement
3)No training phase
4)Works well on small datasets
5)Can solve both classification and regression problems


Disadvantages:

1)Slow for large datasets
2)Requires feature scaling
3)Sensitive to noisy data
4)Choosing the right K is important
5)Performance decreases with many features (curse of dimensionality)


```python
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
import pandas as pd

```

```python
df=pd.read_csv('retail_customer_segmentation.csv')
df.head()
```

```python
df.info()
```

```python
#check null values
df.isnull().sum()
```

```python
#this for single column
df["annual_income"]=df["annual_income"].fillna(df["annual_income"].median())
```

```python
df.isnull().sum()
```

```python
#this for multiple columns  it replaces througth median values with NaN
df.fillna(df.median(numeric_only=True),inplace=True)
```

```python
df.isnull().sum()
```

```python
#separate the data into features and target

x=df.drop("customer_segment",axis=1)

#use label encoding for target col
le=LabelEncoder()
y=le.fit_transform(df["customer_segment"])
```

```python
x.info()
```

Here Still after clearing null values and separeated feature and target  and features still we have object columns.we need to convert to numuric values 

-here unique values are 3 in both we can use LabelEncoder

```python
#we create separete obj for labelencoser
f_le=LabelEncoder()
obj_col=x.select_dtypes(include=object).columns
for col in obj_col:
    x[col]=f_le.fit_transform(x[col])
```

```python
x.info()
```

#Now every column is integer

```python
#split the data 
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
xtrain
```

```python
#use scalling tecqniques-->StandardScaler 

scaler=StandardScaler()
#for train data
x_train_updated=scaler.fit_transform(xtrain)
#for test data we use only transform  
x_test_updated=scaler.transform(xtest)
```

```python
# #creates model for KNN
# for k in range(1,5):
#     model=KNeighborsClassifier(n_neighbors=k)
#     model.fit(x_train_updated,ytrain)

#     y_pred_train=model.predict(x_train_updated)

#     print(f" For k :{k} Accuracy Score:{accuracy_score(ytrain,y_pred_train)}")
    
```

```python
model=KNeighborsClassifier(n_neighbors=5)
model.fit(x_train_updated,ytrain)

y_pred_train=model.predict(x_train_updated)

print(f" For k :{5} Accuracy Score:{accuracy_score(ytrain,y_pred_train)}")
```

```python
print(confusion_matrix(ytrain,y_pred_train))
```

```python
print(classification_report(ytrain,y_pred_train, target_names=le.classes_))
```

```python
#Model evaluation for test data

y_test_pred=model.predict(x_test_updated)

print("Accuracy Score for train data:",accuracy_score(ytest,y_test_pred))
```

```python
print(confusion_matrix(ytest,y_test_pred))
```

```python
print(classification_report(ytest,y_test_pred, target_names=le.classes_))
```

Conclusion

The KNN classifier achieved an accuracy of 62.12% on the test dataset. Among all customer segments, Occasional customers were classified most accurately, achieving an F1-score of 0.72 and a recall of 0.76, indicating that the model successfully identified most customers in this segment. The performance for High_Value and Loyal customers was moderate, with F1-scores of 0.61 each. However, the model struggled to classify Regular customers, resulting in a lower precision of 0.50, recall of 0.40, and F1-score of 0.44. Overall, the model provides moderate classification performance, but improving feature engineering or tuning the value of K may enhance the prediction accuracy, especially for the Regular customer segment.
