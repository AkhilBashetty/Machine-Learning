```python
#Scaling Techqniue
''' 
Scaling---->we use scaling  for scale the values into similler scale values .
 ------>Scaling useually do after spliting train and test data because from train data model should be learn and from test data it should predict
             the data.

             Train data= we use fit_transform()
             Test data=we use just tranform()-->if we use again fit here again fit predict again the new data that value that transform again.
 

---->when we use distance formula  distance has long values with compare to other values we do scaling to get into the same range 

1)StandardScaler
2)MinMaxScaler
3)RobustScaler

'''
```

```python
#1)Min-Max Scaler - we get values with in the range of 0-1
'''  
Min-Max Scaler transforms the data so that all values lie between 0 and 1 (default range).
we use min val  and max val in column


            formula :  X(scale)= x-min_val/max_val-min_val

            Xscaled​=X-Xmin/Xmax​−Xmin


'''
import pandas as pd
import numpy as np
n_rows=15
data={
    "Age":np.random.randint(5,100,n_rows)
}
df=pd.DataFrame(data)
df


#scaling
from sklearn.preprocessing import MinMaxScaler

#creating obj before this we need to split the data for real time project but it just example I am not using train_test_split()
mms=MinMaxScaler()
df["Age_scaled"]=mms.fit_transform(df[["Age"]])
df
```

```python
#2. Standard Scaler:Instead of range like 0-1 it uses mean and standard deviation
'''

            Z=X−μ/σ​
Where

X = Original value
μ = Mean
σ = Standard deviation

            
'''
from sklearn.preprocessing import StandardScaler

ss=StandardScaler()
df["Age_scaled_StandardScaler"]=ss.fit_transform(df[["Age"]])
df
```

```python
# Robust Scaler
'''  
3. Robust Scaler

Robust Scaler uses:

Median
Interquartile Range (IQR)

instead of mean and standard deviation.

Since median is not affected much by extreme values, Robust Scaler handles outliers better.

'''

from sklearn.preprocessing import RobustScaler

#object creation

rs=RobustScaler()

df["Age_scaled_RobustScaler"]=rs.fit_transform(df[["Age"]])
df
```
